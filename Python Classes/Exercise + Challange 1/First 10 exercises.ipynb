###Exercício 01: 




Braian é um professor legal e quer evitar que o aluno João seja reprovado. João tirou 8,66, 5,35, 5 e 1, respectivamente, nas provas P1, P2, P3 e P4. Para isso, ele pode calcular a nota final usando a média aritmética (M.A.), média geométrica (M.G.) ou média harmônica (M.H.) 
Qual dessas médias dá a maior nota para João? E qual das médias dá a pior nota?

```
from statistics import *
import numpy as np


def geo_mean(iterable):
    a = np.array(iterable)
    return a.prod()**(1.0/len(a))


p1 = 8.66
p2 = 5.35
p3 = 5
p4 = 1


provas = [p1,p2,p3,p4]

f1 = mean(provas)
f2 = geo_mean(provas)
f3 = harmonic_mean(provas)
print("Maior")
if f1>=f2:
  if f1>=f3:
    print("MA "+str(f1))
  else:
    print("MH "+str(f3))
elif f2>=f3:
  print("MG "+str(f2))
else:
  print("MH "+str(f3))

print()
print("Menor")

if f1<=f2:
  if f1<=f3:
    print("MA "+str(f1))
  else:
    print("MH "+str(f3))
elif f2<=f3:
  print("MG "+str(f2))
else:
  print("MH "+str(f3))
```

Dada a lista [2, 8, 3, 13, 7, 24, “str”, [1, 2, 11], 8, True], exiba todos os números primos existentes em seu conteúdo em ordem crescente (incluindo a lista interna).




```
import numpy as np

def isPrimo(candidato):
    if(candidato%2 == 0 and candidato !=2): return False;
    for eachValue in range(3,candidato//2+1,2):
        if(candidato%eachValue == 0):
            return False
    return True

#  --------------



lista = [2, 8, 3, 13, 7, 24, 'str', [1, 2, 11], 8, True]
primos = []
for i in lista:
    if hasattr(i, '__iter__'):
        for value in i:
            if(type(value) == int and isPrimo(value)):
                primos.append(value)
    else:
        if(type(i) == int and isPrimo(i)):
            primos.append(i)


print(np.sort(list(set(primos))))
```

Lembrem-se que podem existir listas dentro de listas.


```
import numpy as np

def isPrimo(candidato):
    if(candidato%2 == 0 and candidato !=2): return False;
    for eachValue in range(3,candidato//2+1,2):
        if(candidato%eachValue == 0):
            return False
    return True

#  --------------

def isValidPrimo(value):
    try:
        value = int(value)
        if(type(value) == int and isPrimo(value)):
            return value
    except:
        pass


# ---------------------
def getPrimosFromList(lista):
    primos = []
    for i in lista:
        if (hasattr(i, '__iter__') and i != lista):
            getPrimosFromList(i)
        else:
            primos.append(isValidPrimo(i))

    return (np.sort(list(set(filter(None, primos)))))


lista = [2, 8, 3, 13, 7.5, 24, 'str', [1, 2, 11], 8, True]
getPrimosFromList(lista)
```
