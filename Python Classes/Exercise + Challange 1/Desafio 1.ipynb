Construa um módulo conta_bancaria que possui uma classe Cliente e outra classe Extrato.

A classe Extrato guarda a data de criação da conta e uma lista de mensagens informando as transações já realizadas. Funções: adicionar_transacao e gerar_extrato

A classe Cliente possui os atributos nome, agencia, conta, saldo e extrato (do tipo Extrato) Os atributos são inicializados na criação do objeto; Funções: sacar, depositar, exibir_saldo, transferir e exibir_extrato.

Lembre-se que o usuário não pode gastar mais do que tem.

```python
from datetime import datetime

class Extrato:

    def __init__(self):
        self._create_date=datetime.now()
        self.transactions = []

    def adicionar_transacao(self,new_transaction):
        self.transactions.append(new_transaction)

    def gerar_extrato(self):
        print("***EXTRATO***\n\n")
        for transaction_number,transaction in enumerate(self.transactions,start=1):
          print(str(transaction_number)+" - "+str(transaction))


from IPython.display import clear_output

class Cliente:

    def __init__(self,name,account,agency):
        self.extrato = Extrato()
        self.name = name
        self.account = account
        self.agency = agency
        self.balance = 0

    def sacar(self,quantity):
        clear_output()
        print("")
        assert isinstance(quantity, (int,  float)),"Quantity value bust be a float or int"
        assert 0<=quantity, 'Withdraw quantity must be a number greater than 0'
        assert quantity<=self.balance, 'Cannot Withdraw More than You Current Balance'
        print("PRINTINTING MONEY -> "+("$")*quantity)
        self.balance -= quantity
        self.extrato.adicionar_transacao("Saque de $"+str(quantity))

    def depositar(self,quantity,texto="Deposito"):
        clear_output()
        print("")
        assert isinstance(quantity, (int,  float)),"Quantity value bust be a float or int"
        assert 0<=quantity, 'Deposit quantity must be a number greater than 0'
        print("INSERTING MONEY  <- "+("$")*quantity)
        self.balance += quantity
        self.extrato.adicionar_transacao(str(texto)+" de $"+str(quantity))

    def exibir_saldo(self):
        clear_output()
        print("")
        print("CURRENT MONEY  : "+("$")*self.balance)
        print("CURRENT MONEY  : $"+str(self.balance))
        self.extrato.adicionar_transacao("Exibição de Saldo: $"+str(self.balance))


    def transferir(self,quantity,destination_client):
        clear_output()
        print("")
        assert isinstance(quantity, (int,  float)),"Quantity value bust be a float or int"
        assert 0<=quantity, 'Send quantity must be a number greater than 0'
        assert quantity<=self.balance, "Cannot Send More than You Current Balance"
        try:
            previous_balance = self.balance
            self.balance -= quantity
            destination_client.depositar(quantity,"Transferência recebida de "+self.name)
        except:
            print("***** TRANSACTION FAILED *****")
            self.balance = previous_balance
            return 0
        import time
        for i in range(11):
            clear_output()
            print("SENDING MONEY -> "+("$")*(10-i))
            print("")
            print("RECEVING MONEY <- "+("$")*i)
            time.sleep(0.45)
        self.extrato.adicionar_transacao("Transferência de $"+str(quantity))

        

    def exibir_extrato(self):
        clear_output()
        self.extrato.gerar_extrato()
```

```python
cliente1 = Cliente("Marlene",2,3)
cliente2 = Cliente(1,2,3)


cliente1.depositar(100)

cliente1.transferir(100,cliente2)
```
